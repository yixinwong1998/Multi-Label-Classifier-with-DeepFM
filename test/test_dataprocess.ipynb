```python
import pandas as pd
import numpy as np
```

```python
label = pd.read_excel("ADMET.xlsx")
df = pd.read_excel("Molecular_Descriptor.xlsx")
data = df.iloc[:, 1:df.shape[1]]  # remove the first column with description of molecular
```

```python
def sparse_or_dense(X):
    data = np.array(X)
    sparse_cols = []
    dense_cols = []
    for i in range(data.shape[1]):
        col = data[:, i]
        if any(col % 1 != 0):
            dense_cols.append(i)
        else:
            sparse_cols.append(i)
    return sparse_cols, dense_cols
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
print(data.shape)
fs_selector = VarianceThreshold(threshold=0.0)
fs_selector.fit(X=data)
```

```python
columns_variances = fs_selector.variances_
print(type(columns_variances))
print(columns_variances.shape)
selected_fs = list(filter(lambda X: X != 0, columns_variances))
print(len(selected_fs))
selected_fs_index = [i for i in range(len(columns_variances)) if columns_variances[i] != 0]
print(len(selected_fs_index))
print(selected_fs_index)
print(fs_selector.variances_)
```

```python
data = data.iloc[:,selected_fs_index]
print(data.shape, type(data))
data.head()
```

```python
sparse_cols, dense_cols = sparse_or_dense(data)
print(len(sparse_cols), len(dense_cols))
print(sparse_cols, end="\n")
print(dense_cols)
```

```python
columns = data.columns.values  # ["nAcid",	'ALogP', 'ALogp2', 'AMR'...]
sparse_features = columns[sparse_cols].tolist()
dense_features = columns[dense_cols].tolist()
print(sparse_features, end='\n')
print(dense_features)
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
fsed = sparse_features + dense_features
print(len(fsed))
print(type(fsed))
print(data.shape)
print(type(data))
data[fsed].head()
```

```python
data[sparse_features] = data[sparse_features].fillna('-1', )  # 字符型数据
data[dense_features] = data[dense_features].fillna(0, )
# 1.Label Encoding for sparse features,and do simple Transformation for dense features
for feat in sparse_features:
    lbe = LabelEncoder()
    data[feat] = lbe.fit_transform(data[feat])
mms = MinMaxScaler(feature_range=(0, 1))  # min, max = feature_range
data[dense_features] = mms.fit_transform(data[dense_features])
data.head()
```
